```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/df_model_task1.csv")

# Create Target Variable
df["ClaimFlag"] = np.where(df["TotalClaims"] > 0, 1, 0)

# Select Useful Features
model_features = [
    "SumInsured",
    "CalculatedPremiumPerTerm",
    "RegistrationYear",
    "Cylinders",
    "NumberOfDoors",
    "CustomValueEstimate",
    "TotalPremium",
    "Province",
    "Gender",
    "VehicleType",
    "make",
    "ClaimFlag"
]

ml_df = df[model_features].copy()

# Handle categorical variables
ml_df = pd.get_dummies(ml_df, drop_first=True)

# Save ML-ready dataset in data/processed
ml_df.to_csv("../data/processed/ml_dataset.csv", index=False)

ml_df.head()
```
